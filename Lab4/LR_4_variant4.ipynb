## Лінійна регресія (МНК) та інтерполяція

**Варіант 4**

### Завдання 1. Теорія (коротко)

**Лінійна регресія** моделює залежність:

$$y = \beta_0 + \beta_1 x + \varepsilon$$

Параметри $\beta_0, \beta_1$ оцінюються **методом найменших квадратів (МНК)**: мінімізується сума квадратів відхилень між експериментальними $y_i$ та прогнозом $\hat y_i = \beta_0 + \beta_1 x_i$:

$$S(\beta_0,\beta_1) = \sum_{i=1}^n \left(y_i - (\beta_0 + \beta_1 x_i)\right)^2$$

**Інтерполяція**: для заданих точок $(x_i, y_i)$ будується поліном степеня не вище $n-1$, який точно проходить через усі ці точки. Якщо задано 5 точок, то потрібно поліном степеня 4.

### Завдання 2. Лінійна регресія (МНК) для варіанта 4

Дані з методичної:

- $x$: 2, 4, 6, 8, 10, 12
- $y$: 6.5, 4.4, 3.8, 3.5, 3.1, 3.0

```python
import numpy as np
import matplotlib.pyplot as plt

# Вхідні дані для варіанта 4
x = np.array([2, 4, 6, 8, 10, 12], dtype=float)
y = np.array([6.5, 4.4, 3.8, 3.5, 3.1, 3.0], dtype=float)

n = len(x)

# Оцінювання параметрів лінії МНК (нормальні рівняння)
# beta1 = (n*Σ(xy) - Σx*Σy) / (n*Σ(x^2) - (Σx)^2)
# beta0 = (Σy - beta1*Σx) / n
Sxy = n * np.sum(x * y) - np.sum(x) * np.sum(y)
Sxx = n * np.sum(x**2) - (np.sum(x) ** 2)
beta1 = Sxy / Sxx
beta0 = (np.sum(y) - beta1 * np.sum(x)) / n

# Перевірка: сума квадратів похибок
y_hat = beta0 + beta1 * x
SSE = float(np.sum((y - y_hat) ** 2))

print(f"beta0 = {beta0:.6f}")
print(f"beta1 = {beta1:.6f}")
print(f"SSE = {SSE:.6f}")
```

```python
# Побудова графіка: експериментальні точки і апроксимуюча пряма
xx = np.linspace(x.min(), x.max(), 300)
yy = beta0 + beta1 * xx

plt.figure(figsize=(7, 4.5))
plt.scatter(x, y, s=50, label="Експериментальні точки")
plt.plot(xx, yy, linewidth=2, label="Лінія МНК")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, alpha=0.3)
plt.legend()
plt.title("Лінійна регресія (МНК), варіант 4")

import os
LAB_DIR = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
out_path = os.path.join(LAB_DIR, "LR_4_variant4_regression.png")
plt.savefig(out_path, dpi=200, bbox_inches="tight")
print(f"Графік збережено: {out_path}")
plt.show()
plt.close()
```

### Завдання 3. Інтерполяція поліномом степеню 4

Табличні значення з методичної (5 точок):

- $x$: 0.1, 0.3, 0.4, 0.6, 0.7
- $y$: 3.2, 3.0, 1.0, 1.8, 1.9

Поліном шукаємо у вигляді:

$$P_4(x) = a_0 + a_1 x + a_2 x^2 + a_3 x^3 + a_4 x^4$$

```python
# Вхідні точки для інтерполяції
x_i = np.array([0.1, 0.3, 0.4, 0.6, 0.7], dtype=float)
y_i = np.array([3.2, 3.0, 1.0, 1.8, 1.9], dtype=float)

m = 4  # степінь полінома

# Крок 1: заповнення матриці X (Вандермонда, але у вигляді [x^0, x^1, ..., x^4])
# X * a = y, де a = [a0, a1, a2, a3, a4]^T
X = np.vander(x_i, N=m+1, increasing=True)

# Крок 2: отримання коефіцієнтів інтерполяційного полінома
a = np.linalg.solve(X, y_i)

print("Коефіцієнти P4(x) (increasing powers a0..a4):")
for k, ak in enumerate(a):
    print(f"a{k} = {ak:.10f}")
```

```python
# Крок 3-4: визначення полінома і побудова графіка
def P4(t: float) -> float:
    powers = np.array([t**k for k in range(m+1)], dtype=float)
    return float(np.sum(a * powers))

xx = np.linspace(x_i.min(), x_i.max(), 500)
yy = np.array([P4(t) for t in xx], dtype=float)

plt.figure(figsize=(7, 4.5))
plt.scatter(x_i, y_i, s=55, label="Точки інтерполяції")
plt.plot(xx, yy, linewidth=2, label="Поліном P4(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, alpha=0.3)
plt.legend()
plt.title("Інтерполяція поліномом степеню 4")

import os
LAB_DIR = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
out_path = os.path.join(LAB_DIR, "LR_4_variant4_interpolation_P4.png")
plt.savefig(out_path, dpi=200, bbox_inches="tight")
print(f"Графік збережено: {out_path}")
plt.show()
plt.close()
```

```python
# Крок 5: значення в проміжних точках
x1 = 0.2
x2 = 0.5

print(f"P4({x1}) = {P4(x1):.10f}")
print(f"P4({x2}) = {P4(x2):.10f}")
```
